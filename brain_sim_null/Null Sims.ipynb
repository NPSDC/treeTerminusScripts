```r
simDir <- "/fs/cbcb-scratch/jfan03/treeDE-null"
library(glue)
seedFiles <- sapply(seq(10), function(seed) {
    file.path(simDir, glue(paste("output_seed","",sep="="), {seed}), 
          paste(glue("treeSumExp_seed=", {seed}), "RData", sep ="."))
})
```

```r
for(i in seq_along(seedFiles)) {
    print(paste("Loading RData", seedFiles[i]))
    load(file=seedFiles[i])
    tree <- rowTree(tse)
    l <- length(tree$tip)
    yAll <- computeSizeFactors(tse)
    yAll <- scaleInfReps(yAll)
    yAll <- labelKeep(yAll)
    
    set.seed(10)
    yL <- swish(yAll[1:l,], x="condition")
    hist(mcols(yL)[["pvalue"]])
    
    set.seed(10)
    yInn <- swish(yAll[(l+1):nrow(yAll),], x="condition")
    nullLeaves <- sapply(c(0.01, 0.05, 0.1), function(qval) {
        sum(mcols(yL)[["qvalue"]] <= qval, na.rm=T)
    })
    print("Length of null leaves are")
    print(nullLeaves)
    
    hist(mcols(yInn)[["pvalue"]])
    dep <- node.depth(tree, 2)
    hist(mcols(yInn)[["pvalue"]][dep[dep>=6]-l])
    
    pvals <- c(mcols(yL)[["pvalue"]], mcols(yInn)[["pvalue"]])
    signs <- computeSign(yAll, "condition")
    tA <- list()
    tA <- lapply(c(0.01, 0.05, 0.1), function(alpha) {
        climbMax(pvals, mcols(yAll)[["meanInfRV"]], tree, 0.4, alpha, signs, cores=1)
    })
    print("Length of nodes are")
    print(sapply(tA, length))
    print(sapply(tA, function(nodes) sum(nodes > l)))
}
```
