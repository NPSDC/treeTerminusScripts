```r
setwd('/fs/cbcb-lab/rob/students/noor/Uncertainity/treeTerminusScripts')
suppressPackageStartupMessages(source("tree_helper_function.R"))
suppressPackageStartupMessages(library(ggplot2))
suppressPackageStartupMessages(library(ggpubr))
suppressPackageStartupMessages(library(dplyr))
load("environment/brain_sim_nodtu/mode=gc_bias/mapDf.RData")
suppressPackageStartupMessages(library(latex2exp))
```

```r
colValues <- c("Cons" = "#F8766D",
               "Mean" = "#C49A00",
                "ConsFilt" = "#53B400",
                "ConsFiltES" = "#00C094",
                "AC" = "#00B6EB",
               "Txp" = "#FB61D7",
               "Gene" = "#00A9FF"
              )
ord <- c("Cons", "Mean", "ConsFilt", "ConsFiltES", "AC", "Txp", "Gene")
```

```r
dir <- "environment/brain_sim_nodtu_lowfc/mode=gc_bias/post_type=gibbs_nrep=100_tf=100/comp_trees"
load(file=file.path(dir, "dfIT.RData"))
nn <- c("2"="Height=2",
       "3"="Height=3",
       "4"="Height=4",
       "5"="Height>=5")
dfIT$TypeUp <- factor(dfIT$Type, levels = ord)
dfSumm <- dfIT %>%  filter(depth!=1) %>% 
        group_by(TypeUp, modDep) %>% 
        summarise(n=n(), med = median(MIRV), max = max(log2(MIRV))+1)
options(repr.plot.width=20, repr.plot.height=10)
size <- 18
dfIT %>% filter(depth!=1) %>% ggplot(aes(x=TypeUp, y=log2(MIRV), fill=TypeUp)) + 
    geom_violin(width=1.8) + 
    facet_wrap(~modDep, labeller = labeller(modDep = nn)) + 
    theme_bw() +
    theme(text = element_text(size = size), legend.position = "none",
          axis.text = element_text(size = size),
          strip.text = element_text(size=size),
          strip.background = element_rect(fill = "white", color = "white")) +
          scale_color_manual(values = colValues) +
                    scale_fill_manual(values = colValues) +
    scale_fill_manual(values = colValues) +
    ylab(TeX("$log_{2}$(MIRV)")) +
    xlab("Type") +
    geom_text(data=dfSumm, aes(x = TypeUp, y = max, label=n), color="black", fontface = 2, size = 6)
```
