```r
colValues <- c("Cons" = "#F8766D",
               "Mean" = "#C49A00",
                "ConsFilt" = "#53B400",
                "ConsFiltES" = "#00C094",
                "AC" = "#00B6EB",
                "Txp" = "#FB61D7"
              )
ord <- c("Cons", "Mean", "ConsFilt", "ConsFiltES", "AC", "Txp")
```

```r
dir <- "environment/brain_sim_nodtu/mode=gc_bias/post_type=gibbs_nrep=100_tf=100/comp_trees"
load(file.path(dir, "dfInf.RData"))
size <- 20
dfInf <- dfInf %>% filter(gamma!=10)
dfInf$Type[dfInf$Type=="Txps"] = "Txp"
dfInf$gamma <- as.character(dfInf$gamma)
dfInf$TypeUp <- factor(dfInf$Type, levels = ord)
pBrNorm <- ggplot(dfInf, aes(x=gamma, fill = TypeUp, colour = TypeUp, y = Value)) + 
    geom_bar(stat="identity", position = "dodge", alpha=0.7) +
    theme_bw() +
    labs(fill="TypeUp", color="TypeUp") +
    xlab(TeX("$gamma")) +
    ylab("Optimum Value") +
    scale_y_continuous(trans='log10') +
    coord_cartesian(ylim = c(6*1e3, 2*1e6)) +
    ggtitle("BrSimNorm") +
    guides(fill=guide_legend(title="Type"), color=guide_legend(title="Type")) +
    theme(axis.title = element_text(size = size),
    plot.title = element_text(size = size + 1, hjust=0.5),
    axis.text = element_text(size = size),
    legend.text = element_text(size = size),
    legend.title = element_text(size = size)) +
    scale_color_manual(values = colValues) +
    scale_fill_manual(values = colValues)

pBrNormCut <- ggplot(dfInf, aes(x=gamma, fill = TypeUp, colour = TypeUp, y = nCuts)) + 
    geom_bar(stat="identity", position = "dodge", alpha=0.7) +
    theme_bw() +
    labs(fill="Type", color="Type") +
    xlab(TeX("$gamma")) +
    ylab("Cut Size") +
#     scale_y_continuous(trans='log10') +
#     coord_cartesian(ylim = c(2*1e4, 1.4*1e5)) +
    ggtitle("BrSimNorm") +
    guides(fill=guide_legend(title="Type"), color=guide_legend(title="Type")) +
    theme(axis.title = element_text(size = size),
    plot.title = element_text(size = size + 1, hjust=0.5),
    axis.text = element_text(size = size),
    legend.text = element_text(size = size),
    legend.title = element_text(size = size)) +
    scale_color_manual(values = colValues) +
    scale_fill_manual(values = colValues)
```
