```r
setwd('/fs/cbcb-lab/rob/students/noor/Uncertainity/treeTerminusScripts')
suppressPackageStartupMessages(library(ggplot2))
suppressPackageStartupMessages(library(ggpubr))
suppressPackageStartupMessages(library(dplyr))
dir <- "environment/Chimp/mode=gcbias/post_type=gibbs_nrep=100_tf=100/tree_analysis"
load(file.path(dir, "dfFamChimp.RData"))
```

```r
colValues <- c("Cons" = "#F8766D",
               "Mean" = "#C49A00",
                "ConsFilt" = "#53B400",
                "ConsFiltES" = "#00C094",
                "AC" = "#00B6EB"
              )
ord <- c("Cons", "Mean", "ConsFilt", "ConsFiltES", "AC")
```

```r
options(repr.plot.width=10, repr.plot.height=8)

dfFamChimp$TypeUp <-  factor(dfFamChimp$Type, levels = ord)
pChimp <- ggplot(dfFamChimp, aes(x=lab, y=nFam, fill=TypeUp)) + 
    geom_bar(stat="identity", position="dodge", width = 0.2) + 
    scale_y_continuous(trans='log10', expand = c(0,0)) + coord_cartesian(ylim = c(1, 1e5)) +
    scale_x_continuous(trans='log10') +
    theme_bw() +
    theme(text = element_text(size = 20), plot.title = element_text(size=20, hjust = 0.5)) +
    labs(fill="Type", color="Type") +
    guides(fill=guide_legend(title="Type"), color=guide_legend(title="Type")) +
    xlab("Number of Families") +
    ylab("Number of Nodes") + 
    ggtitle("ChimpBrain Dataset") +
    scale_color_manual(values = colValues) +
    scale_fill_manual(values = colValues)
```
