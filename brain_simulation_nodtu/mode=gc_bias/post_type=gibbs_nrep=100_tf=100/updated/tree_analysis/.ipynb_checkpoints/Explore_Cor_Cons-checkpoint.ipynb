```r
nL <- length(trMeanCorUPGMA$tip)
depth <- node.depth(trMeanCorUPGMA,2)
modDepCorr <- depth
modDepCorr[depth >=7] = 7
innNodesCorr <- nL+1:trMeanCorUPGMA$Nnode
childNodesCorr <- Descendants(trMeanCorUPGMA, innNodesCorr, "child")

infRepIdx <- grep("infRep",assayNames(yCorMeanAggS))
infReps <- assays(yCorMeanAggS)[infRepIdx]
infReps <- abind::abind(as.list(infReps), along=3)
infMCorr <- rowMeans(infReps[,,])
```

```r
meanDiffCorr <- sapply(seq_along(innNodesCorr), function(i) {
    cM <- infMCorr[childNodesCorr[[i]]]
    (infMCorr[innNodesCorr[i]]-min(cM))/infMCorr[innNodesCorr[i]]
})
```

```r
dfInf <- data.frame(type="Correlation", depth=modDepCorr[nL+1:trMeanCorUPGMA$Nnode], infRV = mcols(yCorMeanAggS)[nL+1:trMeanCorUPGMA$Nnode,1],
                   meanDiff = meanDiffCorr, expCat = sapply(infMCorr[nL+1:trMeanCorUPGMA$Nnode], function(x) {
                       if(x < 10)
                           return("10")
                       else if(x < 100)
                           return("100")
                       else if(x < 1000)
                           return("1000")
                       else if(x < 10000)
                           return("10000")
                       else
                           return("everything else")
                   }))
```

```r

table(dfInf[["expCat"]])
table(dfInf[["expCat"]],dfInf[["depth"]])
```

```r
red <- sapply(seq_along(innNodes), function(i) mean(mcols(yCorMeanAggS)[childNodes[[i]],1] > mcols(yCorMeanAggS)[innNodes[i],1]))
```

```r
redInfRV <- sapply(seq_along(innNodesCorr), function(i) mcols(yCorMeanAggS)[innNodesCorr[i],1] - mean(mcols(yCorMeanAggS)[childNodesCorr[[i]],1]))
```

```r
length(redInfRV)
summary(redInfRV)
mean(redInfRV < 0)
summary(redInfRV[redInfRV < 0])
```

```r
dfRed <- data.frame(depth=modDep[modDep > 1], mProp=red)
pL <- list()
for(i in c(2:7)) {
    dfR <- dfRed[dfRed[["depth"]] == i, ]
    

    pL[[i-1]] <- ggplot(dfR, aes(x=mProp)) +  geom_histogram()     #+facet_wrap(vars(modDep))
}
ggarrange(plotlist=pL)
```
