```r
getInpForHittingSet <- function(genes, geneMap, emptyGeneSet) {
    genes <- setdiff(genes, emptyGeneSet)
    fams <- unique(unlist(geneMap[genes]))
    geneFamsList <- geneMap[genes]
    gsets <- sapply(geneFamsList, function(fam) paste(match(fam, fams), collapse=" "))
    return(gsets)

}
```

```r
length(corMultInd)
for(i in corMultInd) {
    print(i)
    genes <- genesCor[[corMultInd[i]]]
    gsets <- getInpForHittingSet(genes, geneFam, geneEmptyFam)
    write(">>>", append=T)
    for(j in seq_along(gsets)) {
        write(paste(gsets[[j]], collapse=" "), file.path(gene_dir, "corHitInp0.txt"), append=T)
    }
}
```

```r
tt <- rep(0, length(unique(unlist(gsets))))
names(tt) <- as.character(unique(unlist(gsets)))
gs <- lapply(gsets, function(sets) {
    remSets <- sets[tt[as.character(sets)] != 0]
    if(length(remSets)!=0) {
        sets <- setdiff(sets)
    }
    if(sum(tt[sets]==1)!=length(sets))
        sets <- sets[tt[sets]!=1]
    tt[as.character(sets)] += 1
    sets
})
```
